```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import pandas as pd

# Define the pattern
pattern_entity = r'ET: {<NNP>+<IN>?<NNP>+|<NNP>+}'

# Define a function to preprocess text and apply the pattern
def extract_patterns(text):
    sent = nltk.word_tokenize(text)
    sent = nltk.pos_tag(sent)
    cp = nltk.RegexpParser(pattern_entity)
    cs = cp.parse(sent)
    patterns = []
    for subtree in cs.subtrees():
        if subtree.label() == 'ET':
            patterns.append(' '.join(word for word, pos in subtree.leaves()))
    return patterns

# Apply the function to each row of the 'Texts' column and store the patterns in a new column
filtered_df['extracted_patterns'] = filtered_df['Texts'].apply(extract_patterns)
```

```python
import pandas as pd
from collections import Counter

# Flatten the list of lists in the 'extracted_patterns' column
all_words = [word for sublist in filtered_df['extracted_patterns'] for word in sublist]

# Count the frequency of each word
word_freq = Counter(all_words)

# Convert the word frequency counter to a DataFrame
word_freq_df = pd.DataFrame.from_dict(word_freq, orient='index', columns=['Frequency'])

# Sort the DataFrame by frequency in descending order
word_freq_df = word_freq_df.sort_values(by='Frequency', ascending=False)

# Print the word frequency DataFrame
print(word_freq_df)
```

```python
word_freq_df = word_freq_df.to_csv('frequency.csv')
```

```python
word_freq_df = pd.read_csv('frequency.csv')
word_freq_df.rename(columns={'Unnamed: 0': 'Entities'}, inplace=True)
column_list = word_freq_df['Entities'].tolist()
```

```python
import spacy

# Load the English language model
nlp = spacy.load('en_core_web_sm')

# Function to classify entities
def classify_entities(words):
    entities = {}
    text = ','.join(words)  # Join the words into a single string
    print(text)
    doc = nlp(text)
    for ent in doc.ents:
        entities[ent.text] = ent.label_
    return entities

# Classify the entities
classified_entities = classify_entities(column_list)
print(classified_entities)
```

```python

# Create a new dictionary to store the corrected key-value pairs
corrected_data = {}

for key, value in classified_entities.items():
    if ',' in key:
        # If the key contains commas, split it and assign the same value for each split key
        keys = key.split(',')
        for k in keys:
            corrected_data[k.strip()] = value
    else:
        # If the key does not contain commas, simply assign the value to the key
        corrected_data[key] = value

print(corrected_data)
```

```python
# Define mapping for the values to be updated
mapping = {'GPE': 'LOCATION', 'FAC': 'LOCATION', 'LOC': 'LOCATION', 'PERSON': 'PARTIES', 'ORG': 'PARTIES', 'PRODUCT': 'PARTIES', 'NORP': 'PARTIES'}

# Iterate through the dictionary and update the values
for key, value in corrected_data.items():
    if ',' in key:
        keys_to_update = key.split(',')
        for k in keys_to_update:
            corrected_data[k] = mapping.get(value, value)
        del corrected_data[key]
    else:
        corrected_data[key] = mapping.get(value, value)

# Filter and print entries with values 'LOC' or 'ENTITY'
filtered_data = {k: v for k, v in corrected_data.items() if v in ['LOCATION', 'PARTIES']}
print("Filtered dictionary:", filtered_data)
```

```python
# Create a new column for category in word_freq_df
word_freq_df['Category'] = ''

# Iterate over each row in word_freq_df
for idx, row in word_freq_df.iterrows():
    word = row['Entities']
    # Search for the word in filtered_data keys
    if word in filtered_data:
        category = filtered_data[word]
        word_freq_df.at[idx, 'Category'] = category

print(word_freq_df)
```

```python
import matplotlib.pyplot as plt

# Assuming word_freq_df contains the word frequencies with the 'Category' column

# Filter the DataFrame for locations and parties
locations_df = word_freq_df[word_freq_df['Category'] == 'LOCATION'].head(10)
print(locations_df)
parties_df = word_freq_df[word_freq_df['Category'] == 'PARTIES'].head(10)
print(parties_df)
# Plot the bar graph for locations
plt.figure(figsize=(10, 6))
plt.bar(locations_df['Entities'], locations_df['Frequency'], color='#BF6535')
plt.xlabel('Locations')
plt.ylabel('Frequency')
plt.title('Top 10 Frequencies for Locations 2015-2019')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot the bar graph for parties
plt.figure(figsize=(10, 6))
plt.bar(parties_df['Entities'], parties_df['Frequency'], color='#99A88C')
plt.xlabel('Parties')
plt.ylabel('Frequency')
plt.title('Top 10 Frequencies for Parties 2015-2019')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from datetime import datetime
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import matplotlib.pyplot as plt

# Create dataframe
df = pd.read_csv('Leather_Mag_Text_Cleaned.csv')
df.drop(columns=['Unnamed: 0'], inplace=True)

# Remove \n
for column in df.columns:
    df[column] = df[column].replace('\n', '', regex=True)

# Split the "TitleAndDate" column into words and select the last three words
df['TitleAndDate'] = df['TitleAndDate'].str.split().apply(lambda x: ' '.join(x[-3:]))
df['TitleAndDate'] = pd.to_datetime(df['TitleAndDate'], format='%d %B %Y')

# Sort the DataFrame by the 'TitleAndDate' column
df_sorted = df.sort_values(by='TitleAndDate')

# Define the pattern for extracting event phrases
pattern_entity = r'EV: {<NNP>+<.*>*<VB.*><.*>*<NNP>+}'

# Define a function to preprocess text and apply the pattern
def extract_patterns(text):
    sent = nltk.word_tokenize(text)
    sent = nltk.pos_tag(sent)
    cp = nltk.RegexpParser(pattern_entity)
    cs = cp.parse(sent)
    patterns = []
    for subtree in cs.subtrees():
        if subtree.label() == 'EV':
            patterns.append(' '.join(word for word, pos in subtree.leaves()))
    return patterns

# Apply the function to each row of the 'Title' column and store the patterns in a new column
df_sorted['extracted_events'] = df_sorted['Title'].apply(extract_patterns)
non_empty_events_df = df_sorted[df_sorted['extracted_events'].apply(lambda x: len(x) > 0)]
# Create a dictionary from non_empty_events_df
events_dict = dict(zip(non_empty_events_df['extracted_events'].apply(tuple), non_empty_events_df['TitleAndDate']))

# Print the dictionary
print(events_dict)

import pandas as pd

# Convert the dictionary to a DataFrame
events_df = pd.DataFrame(events_dict.items(), columns=['Extracted Events', 'TitleAndDate'])

# Print the DataFrame
print(events_df)
events_df.to_csv('events.csv')
# Create a dictionary to store the events by year
events_by_year = {}
print(events_by_year)

# Iterate through each row to extract events and their corresponding year
for index, row in df_sorted.iterrows():
    events = row['extracted_events']
    year = row['TitleAndDate'].year
    if events:
        if year not in events_by_year:
            events_by_year[year] = []
        events_by_year[year].extend(events)

# Plotting the timeline
plt.figure(figsize=(12, 6))
for year, events in events_by_year.items():
    plt.plot([datetime(year, 1, 1)] * len(events), range(len(events)), 'o', label=str(year))

plt.xlabel('Year')
plt.ylabel('Event Count')
plt.title('Timeline of Events')
plt.legend()
plt.grid(True)
plt.show()
```
